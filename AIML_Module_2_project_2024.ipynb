```python
## we will be comparing rest of the parameters/columns present in the data with respect to precence or absece of heart disease
data['target'] = data.target.replace({1: "Disease", 0: "No_disease"})
data['sex'] = data.sex.replace({1: "Male", 0: "Female"})
data['cp'] = data.cp.replace({1: "typical_angina",
                          2: "atypical_angina",
                          3:"non-anginal pain",
                          4: "asymtomatic"})
data['exang'] = data.exang.replace({1: "Yes", 0: "No"})
data['slope'] = data.cp.replace({1: "upsloping",
                          2: "flat",
                          3:"downsloping"})
data['thal'] = data.thal.replace({1: "fixed_defect", 2: "reversable_defect", 3:"normal"})
```

cleaning and filling the missing values in the data

```python
data['Caffeine (mg)'] = data['Caffeine (mg)'].replace('Varies', np.NaN).replace('varies', np.NaN)
data['Caffeine (mg)'] = data['Caffeine (mg)'].astype(np.float64)
data['Caffeine (mg)'] = data['Caffeine (mg)'].fillna(data['Caffeine (mg)'].mean())
```

```python
data['Total Fat (g)'].unique()
```

```python
data['Total Fat (g)'] = data['Total Fat (g)'].replace('3 2', '3.2')
```

```python
data.info()
```

```python
# Extract columns with int and float types
numeric_columns = data.select_dtypes(include=["int64", "float64"]).columns

# Print the numeric columns
print(numeric_columns)
```

We will be analysing the dataset using the fact that weather the drink comes under the category tea or not

```python
data['Beverage_category'].unique()
```

```python
data['Tea'] = data['Beverage_category'].apply(lambda x: 1 if x == 'Tazo® Tea Drinks' else 0)
data = data.drop('Beverage_category', axis=1)
```

```python
##  one hot encoding of categorical features in data
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
```

```python
data = onehot_encode(data,columns=['Beverage', 'Beverage_prep'],prefixes=['bev', 'bevp'])
```

```python
data = data.replace({True: 1, False: 0})
```

```python
data = data.applymap(lambda x: np.float64(str(x).replace('%', '')))
```

```python
data.head()
```

```python
data.info()
```

```python
## to do

# Create a pie chart of the 'Tea' column also write your observation form the plot
tea_counts = data['Tea'].value_counts()

# pie chart
plt.pie(tea_counts, labels=['Non-Tea', 'Tea'], autopct='%1.1f%%', startangle=90, colors=['blue', 'lightgreen'] )
plt.title('Proportion of Tea and Non-Tea Drinks in Starbucks Menu')
plt.show()

# Observations
print("Observations:")
print("- Based on the pie chart, it appears that the majority of Starbucks beverages are categorized as Non-Tea.")
print("- Tea drinks constitute approximately 21.5% of the menu and Non-Tea dirnks constitute approximately 78.5% of the menu.")
```

```python
## to do
# perform pca on the data and plot the explained variace ratio, what is the optimal number of principal components in this case ?

from sklearn.preprocessing import StandardScaler

numeric_data = data.select_dtypes(include=['number']).drop(columns=['Tea'])
x = StandardScaler().fit_transform(numeric_data)

pca = PCA().fit(x)
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance Ratio')
plt.show()

n_components_95 = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= 0.95) + 1
print(f"Optimal number of components (95% variance): {n_components_95}")
```

```python
## to do
# visualise the principal components, choose the number of principal components based on the above plot. What is you observation from the plot?

n_components = n_components_95  # From the previous PCA explained variance calculation

pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(x)

# Creating a DataFrame with principal components and the 'Tea' column
pca_df = pd.DataFrame(data=principal_components, columns=[f'PC{i+1}' for i in range(n_components)])
pca_df['Tea'] = data['Tea']

sns.scatterplot(x='PC1', y='PC2', hue='Tea', data=pca_df)
plt.title('PCA Visualization of Starbucks Drinks')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Observations
print("Observations:")
print("- There is a partial separation between tea and non-tea drinks.")
print("- Non-tea drinks occupy a broader area in the PCA space, while tea drinks are relatively clustered.")
```

```python
## to do
# plot the first 2 components of tsne, whats you observation from the plot?
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2)
tsne_data = tsne.fit_transform(data)

# Creating a dataframe
tsne_df = pd.DataFrame({
       "tsne_1": tsne_data[:, 0],
       "tsne_2": tsne_data[:, 1],
       "Tea": data["Tea"] })

sns.scatterplot(x="tsne_1", y="tsne_2", hue="Tea", data=tsne_df)
plt.title("t-SNE Visualization of Starbucks Drinks")
plt.show()

# Observation
print("Observations")
print("- Partial Separation: Tea and non-tea drinks show partial but incomplete separation.")
print("- Tea Cluster: Tea drinks are relatively clustered, indicating greater feature similarity.")
print("- Non-Tea Diversity: Non-tea beverages are more spread out, indicating varying nutritional compositions.")
```

```python
## to do
# create a correlation matrix and plot the heatmap, whats your observation from the heatmap ?

correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()

# Observations from the heatmap (example)
print("Observations:")
print("- There is a strong positive correlation between 'Calories' and 'Total Fat (g)'")
print("- 'Sugars (g)' and 'Carbohydrates (g)' are also positively correlated.")
print("- 'Fiber (g)' and 'Protein (g)' show a moderate positive correlation.")
```

```python
## to do
# make a boxplot of all the numeric columns of the dataset. Which column/columns can be the most potential indicator weather its a tea or a non tea drink?

numeric_cols = data.select_dtypes(include=['number']).drop(columns=['Tea'])

for col in numeric_cols:
    sns.boxplot(x='Tea', y=col, data=data)
    plt.title(f'{col} vs. Tea')
    plt.show()

# Observations
print("Observations:")
print("- Caffeine (mg): Tea beverages contain a relatively high amount of caffeine.")
print("- Calories: Usually, non-tea drinks contain higher calories.")
print("- Total Fat (g): Non-tea drinks tend to have higher total fat.")
print("- Sugars (g): Many non-tea beverages have much greater sugar content.")
```
